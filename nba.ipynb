```python
def height_to_centemeters(height):
    feet, inches = map(int,height.split("-"))
    height_m=feet*0.3048+inches*0.0254
    height_to_centemeters= height_m*100
    return height_to_centemeters
```

```python
def height_to_meters(height):
    try:
        # Ensure the value is a string and split it
        feet, inches = map(int, str(height).split('-'))  
        return feet * 0.3048 + inches * 0.0254  # Convert to meters
    except ValueError:
        return None  # Return None for invalid or missing values
 
# Apply the function to the Height column
df['Height_meters'] = df['Height'].apply(height_to_meters)
 
# Check the updated DataFrame
print("Dataset with numeric Height (in meters):")
print(df[['Name', 'Height', 'Height_meters']])
```

```python
# Function to convert weight from lbs to kg
def weight_to_kg(weight_lbs):
    return weight_lbs * 0.453592

# Function to calculate BMI
def calculate_bmi(row):
    weight_kg = weight_to_kg(row['Weight'])
    height_m = row['Height_meters']
    bmi = weight_kg / (height_m ** 2)
    return bmi

# Add a new column 'BMI' based on existing columns
df['BMI'] = df.apply(calculate_bmi, axis=1)

# Print the updated DataFrame
df
```
